```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange
from scipy import interpolate
```

```python
def R(x):
    return 1/(1+25*x**2)
```

```python
x = np.linspace(-1,1,1000)
```

```python
plt.plot(x,R(x))
```

```python
N = [10]
```

```python
fig1 = plt.plot()
xs = np.linspace(-1,1,1000)
ys = R(xs)
plt.plot(xs,ys)

for n in N:
    
    x = np.zeros(n+1)
    
    #Creating equally spaced pts.
    for i in range(n+1):
        x[i] = (2*i/n) - 1
    
    y = R(x) #Array of Runge function values at the points
    
    poly = lagrange(x,y)
    
    
    #Finding the error 
    
    fs = 0 
    #Values from the interpolating polynomial 
    for i in range(n+1):
        fs += poly[i] * xs**i
    
    error = sum(np.fabs(ys-fs))
        
    print('For n =',n ,',  error =', error)
    
    plt.plot(xs,fs)
```

```python
f_spline = interpolate.CubicSpline(x, y)
```

```python
xs = np.linspace(-1,1,1000)
ys = R(xs)
plt.plot(xs,ys)
plt.plot(xs,f_spline(xs))
```
